#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
y1 = [x**2 for x in x]
y2 = [math.exp(x) for x in x]

fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, sharey = False, figsize = (10,10))

ax1.plot(x, y1, 'r')
ax2.plot(x, y2, 'b')
ax2.set_yscale('log')
ax2.set_xscale('log')
```

#### As a bonus challenge, try to add a legend to the plot.

```python
y1 = [x**2 for x in x]
y2 = [math.exp(x) for x in x]

fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, sharey = False, figsize = (10,10))

ax1.plot(x, y1, 'r')
ax1.legend("x",loc="upper left")
ax1.set_xscale('linear')
ax1.set_yscale('linear')
ax1.grid(True)
ax1.set_xlabel('x axis')
ax1.set_ylabel('x ** 2')


ax2.plot(x, y2, 'b')
ax2.legend("x", loc = 'upper left')
ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.set_xlabel('x axis')
ax2.set_ylabel('e ** x')
ax2.grid(True)

```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
plt.style.use('ggplot')
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, sharey = False, figsize = (12,4))

# MEDIAN STEPS BY MONTH_ENCODED
# your code here
months = df.groupby(df['Months_encoded'])['Steps'].median()
ax1.legend('Months')
ax1.set_ylabel('Steps #')
ax1.set_xlabel('Month')
ax1.bar(df.Months.unique(), months, label = 'Months', color = 'r')
ax1.set_title('Steps per Day Relative to Month',fontweight = 'bold')
ax1.set_xticklabels(df.Months.unique(), rotation = 90)

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
weekend = df.groupby(df['Work_or_Weekend'])['Steps'].median()
ax2.bar(df.Work_or_Weekend.unique(), weekend, color = ['g','y'])
ax2.set_title('Steps on Weekday vs Weekends', fontweight = 'bold')
ax2.set_ylabel('Steps #')
ax2.locator_params(axis='x', nbins=2)

plt.savefig(r'C:\Users\Gareth\Desktop\Ironhack\Week_Five\lab-matplotlib-seaborn\your-code\Figures\Median steps by monthMedian steps by Work & Weekend.jpg')
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
plt.style.use('ggplot')

for column in df.columns[5:9]:
    if column == 'Minutes Fairly Active':
        continue
    else:
        figure, ax = plt.subplots()
        ax.scatter(df[column], df['Steps'])
        ax.set_title(f"{column} vs # of Steps")
        ax.set_xlabel(column)
        ax.set_ylabel('# of Steps')
        plt.show()
```
